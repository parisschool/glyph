```python
# SECCIأ“N 1: أپRBOL AVL 
from glyph import AVLTree

print("1. Creamos un أ،rbol vacأ­o")
avl = AVLTree()

# Insertar elementos que fuercen rotaciones
print("\n2. Insertando 30, 20 y 10 (Observa la rotaciأ³n automأ،tica):")
for x in [30, 20, 10]:
    avl.insert(x)
# Mostrar el estado actual
avl.show()

# Insertar mأ،s para ver un أ،rbol mأ،s complejo
print("\n3. Insertando mأ،s valores para expandir el أ،rbol:")
for x in [40, 50, 25]:
    avl.insert(x)

avl.show()

# Probar la eliminaciأ³n (Borrar)
print("\n4. Eliminando un nodo con hijos (raأ­z):")
avl.delete(30) 
avl.show()
```

```python
# SECCIأ“N 2: BINARY HEAPS (MinHeap)
from glyph import MinHeap

print("1. Inicializamos un MinHeap vacأ­o")
heap = MinHeap()

# Insertar elementos desordenados
# Al insertar el 5 al final, deberأ­a subir hasta la raأ­z
print("\n2. Insertando elementos: [20, 15, 30, 5]")
for x in [20, 15, 30, 5]:
    print(f"Insertando {x}...")
    heap.insert(x)
    # Mostramos el estado en cada paso para ver cأ³mo flota el menor
    heap.show()

print("\n3. Extrayendo el mأ­nimo (El corazأ³n del Heap)")
# extract_min deberأ­a devolver el 5 y reacomodar el resto
min_val = heap.extract_min()
print(f"Valor extraأ­do: {min_val}")

print("\n4. Estado del Heap despuأ©s de extraer el mأ­nimo:")
heap.show()

print("\n5. Insertando un nuevo valor pequeأ±o (2):")
heap.insert(2)
heap.show()
```

```python
# --- SECCIأ“N 3: SKIPLIST (Caminos Rأ،pidos) ---
from glyph import SkipList
import random

print("1. Inicializamos SkipList (p=0.5)")
sk = SkipList(p=0.5)

# Insertar una secuencia para ver cأ³mo se crean los niveles
print("\n2. Insertando una secuencia numأ©rica [10, 20, 30, 40, 50, 60]")
for x in [10, 20, 30, 40, 50, 60]:
    sk.insert(x)

# Mostramos la estructura (Aquأ­ es donde brillan tus 'puntos' y niveles)
sk.show()

print("\n3. Caso de 'Miss' (Dato no encontrado)")
# Intentamos buscar algo que estأ، entre dos nodos para ver la lأ³gica
print("Buscando el 25...")
found = sk.search(25)
print(f"آ؟Resultado?: {'Encontrado' if found else 'No estأ، en la lista'}")

print("\n4. Borrado y Reestructuraciأ³n")
# Si implementaste delete, vemos cأ³mo los niveles superiores se ajustan
if hasattr(sk, 'delete'):
    print("Borrando el 40 (que probablemente tiene varios niveles)...")
    sk.delete(40)
    sk.show()
else:
    print("Operaciأ³n de borrado no disponible, saltando al siguiente ejemplo.")
```

```python
# --- SECCIأ“N 4: BLOOM FILTER (Filtros Probabilأ­sticos) ---
from glyph import BloomFilter

print("1. Inicializamos un Bloom Filter de tamaأ±o pequeأ±o (m=20)")
# Usamos un tamaأ±o pequeأ±o para que sea fأ،cil visualizar las colisiones de bits
bf = BloomFilter(size=20, num_hashes=3)

print("\n2. Insertando elementos:")
items = ["apple", "google", "itam", "data"]
for item in items:
    print(f"Insertando '{item}'...")
    bf.insert(item)

# Aquأ­ tu mأ©todo show deberأ­a mostrar el arreglo de bits (0s y 1s)
bf.show()

print("\n3. Verificando existencias (Hits vs Misses):")

# Un HIT seguro
check_true = "itam"
print(f"آ؟Estأ، '{check_true}'?: {bf.contains(check_true)} (Deberأ­a ser True)")

# Un MISS seguro
check_false = "microsoft"
print(f"آ؟Estأ، '{check_false}'?: {bf.contains(check_false)} (Deberأ­a ser False)")

print("\n4. El concepto de Falso Positivo:")
print("Nota: Si el filtro dice 'True' para algo que nunca insertamos, es un falso positivo.")
# Como el tamaأ±o es 20, si seguimos insertando, eventualmente los bits se llenarأ،n
# y empezarأ، a decir que todo 'quizأ،s' estأ،.
```

```python
# --- SECCIأ“N 5: HASH TABLE (Manejo de Colisiones) ---
from glyph import HashTable

print("1. Inicializamos una Tabla Hash de tamaأ±o pequeأ±o (m=5)")
# Usamos un tamaأ±o pequeأ±o para forzar que los datos choquen entre sأ­
ht = HashTable(size=5)

print("\n2. Insertando elementos:")
# Elige claves que sepas que van a colisionar segأ؛n tu funciأ³n hash
# Si usas el residuo (key % 5), el 1 y el 6 caerأ،n en el mismo أ­ndice.
data_to_insert = [
    (1, "Paris"), 
    (6, "Adriأ،n"), 
    (2, "ITAM"), 
    (7, "Data Science")
]

for key, val in data_to_insert:
    print(f"Insertando clave {key} con valor '{val}'...")
    ht.insert(key, val)

# Aquأ­ tu mأ©todo show debe mostrar las 'cubetas' (buckets)
# y cأ³mo los elementos colisionados cuelgan de la misma posiciأ³n
ht.show()

print("\n3. Bأ؛squeda con أ©xito (Hit):")
key_to_find = 6
result = ht.search(key_to_find)
print(f"Buscando clave {key_to_find}: {result}")

print("\n4. Bأ؛squeda de elemento inexistente (Miss):")
key_miss = 10
result_miss = ht.search(key_miss)
print(f"Buscando clave {key_miss}: {'Encontrado' if result_miss else 'No encontrado'}")

if hasattr(ht, 'delete'):
    print("\n5. Eliminando una colisiأ³n (clave 1):")
    ht.delete(1)
    ht.show()
```

```python
# --- SECCIأ“N 6: STACK (Pila) ---
from glyph import Stack

print("--- Demostraciأ³n de Stack (LIFO) ---")
pila = Stack()

# Push: Meter a la pila
print("1. Agregando elementos (Push): A, B, C")
for item in ['A', 'B', 'C']:
    pila.push(item)
    pila.show()

# Peek: Ver el de arriba sin sacarlo
if hasattr(pila, 'peek'):
    print(f"\n2. Elemento en la cima (Peek): {pila.peek()}")

# Pop: Sacar el أ؛ltimo
print("\n3. Sacando elementos (Pop):")
print(f"Saliأ³: {pila.pop()} (Deberأ­a ser 'C')")
pila.show()
```

```python
# --- SECCIأ“N 7: QUEUE (Cola) ---
from glyph import Queue

print("--- Demostraciأ³n de Queue (FIFO) ---")
cola = Queue()

# Enqueue: Formar en la cola
print("1. Agregando elementos (Enqueue): 1, 2, 3")
for i in range(1, 4):
    cola.enqueue(i)
    cola.show()

# Dequeue: Atender al primero
print("\n2. Atendiendo al primero (Dequeue):")
print(f"Saliأ³: {cola.dequeue()} (Deberأ­a ser '1')")
cola.show()

# Estado Final
print("\n3. Estado final de la cola:")
cola.show()
```

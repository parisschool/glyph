```python
# SECCIÓN 1: ÁRBOL AVL 
from glyph import AVLTree

print("1. Creamos un árbol vacío")
avl = AVLTree()

# Insertar elementos que fuercen rotaciones
print("\n2. Insertando 30, 20 y 10 (Observa la rotación automática):")
for x in [30, 20, 10]:
    avl.insert(x)
# Mostrar el estado actual
avl.show()

# Insertar más para ver un árbol más complejo
print("\n3. Insertando más valores para expandir el árbol:")
for x in [40, 50, 25]:
    avl.insert(x)

avl.show()

# Probar la eliminación (Borrar)
print("\n4. Eliminando un nodo con hijos (raíz):")
avl.delete(30) 
avl.show()
```

```python
# SECCIÓN 2: BINARY HEAPS (MinHeap)
from glyph import MinHeap

print("1. Inicializamos un MinHeap vacío")
heap = MinHeap()

# Insertar elementos desordenados
# Al insertar el 5 al final, debería subir hasta la raíz
print("\n2. Insertando elementos: [20, 15, 30, 5]")
for x in [20, 15, 30, 5]:
    print(f"Insertando {x}...")
    heap.insert(x)
    # Mostramos el estado en cada paso para ver cómo flota el menor
    heap.show()

print("\n3. Extrayendo el mínimo (El corazón del Heap)")
# extract_min debería devolver el 5 y reacomodar el resto
min_val = heap.extract_min()
print(f"Valor extraído: {min_val}")

print("\n4. Estado del Heap después de extraer el mínimo:")
heap.show()

print("\n5. Insertando un nuevo valor pequeño (2):")
heap.insert(2)
heap.show()
```

```python
# --- SECCIÓN 3: SKIPLIST (Caminos Rápidos) ---
from glyph import SkipList
import random

print("1. Inicializamos SkipList (p=0.5)")
sk = SkipList(p=0.5)

# Insertar una secuencia para ver cómo se crean los niveles
print("\n2. Insertando una secuencia numérica [10, 20, 30, 40, 50, 60]")
for x in [10, 20, 30, 40, 50, 60]:
    sk.insert(x)

# Mostramos la estructura (Aquí es donde brillan tus 'puntos' y niveles)
sk.show()

print("\n3. Caso de 'Miss' (Dato no encontrado)")
# Intentamos buscar algo que está entre dos nodos para ver la lógica
print("Buscando el 25...")
found = sk.search(25)
print(f"¿Resultado?: {'Encontrado' if found else 'No está en la lista'}")

print("\n4. Borrado y Reestructuración")
# Si implementaste delete, vemos cómo los niveles superiores se ajustan
if hasattr(sk, 'delete'):
    print("Borrando el 40 (que probablemente tiene varios niveles)...")
    sk.delete(40)
    sk.show()
else:
    print("Operación de borrado no disponible, saltando al siguiente ejemplo.")
```

```python
# Prueba de muchos números
large_sk = SkipList(p=0.5)
for _ in range(20):
    large_sk.insert(random.randint(1, 100))
large_sk.show()
```

```python
# --- SECCIÓN 4: BLOOM FILTER (Filtros Probabilísticos) ---
from glyph import BloomFilter

print("1. Inicializamos un Bloom Filter de tamaño pequeño (m=20)")
# Usamos un tamaño pequeño para que sea fácil visualizar las colisiones de bits
bf = BloomFilter(size=20, num_hashes=3)

print("\n2. Insertando elementos:")
items = ["apple", "google", "itam", "data"]
for item in items:
    print(f"Insertando '{item}'...")
    bf.insert(item)

# Aquí tu método show debería mostrar el arreglo de bits (0s y 1s)
bf.show()

print("\n3. Verificando existencias (Hits vs Misses):")

# Un HIT seguro
check_true = "itam"
print(f"¿Está '{check_true}'?: {bf.contains(check_true)} (Debería ser True)")

# Un MISS seguro
check_false = "microsoft"
print(f"¿Está '{check_false}'?: {bf.contains(check_false)} (Debería ser False)")

print("\n4. El concepto de Falso Positivo:")
print("Nota: Si el filtro dice 'True' para algo que nunca insertamos, es un falso positivo.")
# Como el tamaño es 20, si seguimos insertando, eventualmente los bits se llenarán
# y empezará a decir que todo 'quizás' está.
```

```python
# --- SECCIÓN 5: HASH TABLE (Manejo de Colisiones) ---
from glyph import HashTable

print("1. Inicializamos una Tabla Hash de tamaño pequeño (m=5)")
# Usamos un tamaño pequeño para forzar que los datos choquen entre sí
ht = HashTable(size=5)

print("\n2. Insertando elementos:")
# Elige claves que sepas que van a colisionar según tu función hash
# Si usas el residuo (key % 5), el 1 y el 6 caerán en el mismo índice.
data_to_insert = [
    (1, "Paris"), 
    (6, "Adrián"), 
    (2, "ITAM"), 
    (7, "Data Science")
]

for key, val in data_to_insert:
    print(f"Insertando clave {key} con valor '{val}'...")
    ht.insert(key, val)

# Aquí tu método show debe mostrar las 'cubetas' (buckets)
# y cómo los elementos colisionados cuelgan de la misma posición
ht.show()

print("\n3. Búsqueda con éxito (Hit):")
key_to_find = 6
result = ht.search(key_to_find)
print(f"Buscando clave {key_to_find}: {result}")

print("\n4. Búsqueda de elemento inexistente (Miss):")
key_miss = 10
result_miss = ht.search(key_miss)
print(f"Buscando clave {key_miss}: {'Encontrado' if result_miss else 'No encontrado'}")

if hasattr(ht, 'delete'):
    print("\n5. Eliminando una colisión (clave 1):")
    ht.delete(1)
    ht.show()
```

```python
# --- SECCIÓN 6: STACK (Pila) ---
from glyph import Stack

print("--- Demostración de Stack (LIFO) ---")
pila = Stack()

# Push: Meter a la pila
print("1. Agregando elementos (Push): A, B, C")
for item in ['A', 'B', 'C']:
    pila.push(item)
    pila.show()

# Peek: Ver el de arriba sin sacarlo
if hasattr(pila, 'peek'):
    print(f"\n2. Elemento en la cima (Peek): {pila.peek()}")

# Pop: Sacar el último
print("\n3. Sacando elementos (Pop):")
print(f"Salió: {pila.pop()} (Debería ser 'C')")
pila.show()
```

```python
# --- SECCIÓN 7: QUEUE (Cola) ---
from glyph import Queue

print("--- Demostración de Queue (FIFO) ---")
cola = Queue()

# Enqueue: Formar en la cola
print("1. Agregando elementos (Enqueue): 1, 2, 3")
for i in range(1, 4):
    cola.enqueue(i)
    cola.show()

# Dequeue: Atender al primero
print("\n2. Atendiendo al primero (Dequeue):")
print(f"Salió: {cola.dequeue()} (Debería ser '1')")
cola.show()

# Estado Final
print("\n3. Estado final de la cola:")
cola.show()
```
